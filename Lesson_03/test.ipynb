```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Set the number of DataFrames and their dimensions
num_dataframes = 3
rows, cols = 5, 3  # You can change these values as needed

# Create a loop to generate and display DataFrames
for i in range(num_dataframes):
    # Generate random data
    data = np.random.rand(rows, cols)
    
    # Create a DataFrame
    df = pd.DataFrame(data, columns=[f'Column {j+1}' for j in range(cols)])
    
    # Display the DataFrame
    print(f"DataFrame {i + 1}:")
    display(df)  # Use display function for better output in Jupyter
    plt.pause(1)  # Optional: Pause for 1 second before showing the next DataFrame

```
