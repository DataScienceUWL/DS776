### **Chapter 1: The Mechanics of Learning**

**1. Why is it important to use gradient descent in deep learning models?**  
a) It helps measure the accuracy of the model  
b) It updates model parameters to minimize the loss function  
c) It determines how many layers a model should have  
d) It normalizes the input data for the model  

**Correct Answer: b) It updates model parameters to minimize the loss function**  
*Explanation:* Gradient descent is an optimization technique that iteratively adjusts the model’s weights to reduce the loss function, improving the model’s predictions over time.

### **Chapter 2: Fully Connected Networks**

**6. Consider a fully connected neural network with the following architecture:**  
- Input layer: 10 neurons  
- First hidden layer: 5 neurons  
- Output layer: 2 neurons  

How many parameters (weights and biases) are in the network?  
a) 50  
b) 67  
c) 72  
d) 62  

**Correct Answer: b) 67**  
*Explanation:*  
- Weights from input to hidden layer: \(10 \times 5 = 50\)  
- Biases for hidden layer: \(5\)  
- Weights from hidden to output layer: \(5 \times 2 = 10\)  
- Biases for output layer: \(2\)  
Total: \(50 + 5 + 10 + 2 = 67\).

**7. What is the primary role of the activation function in a fully connected neural network?**  
a) To add additional layers to the network  
b) To introduce non-linearity, allowing the network to learn complex patterns  
c) To initialize the network weights  
d) To determine the number of neurons in each layer  

**Correct Answer: b) To introduce non-linearity, allowing the network to learn complex patterns**  
*Explanation:* Activation functions like ReLU or sigmoid introduce non-linearity, enabling the network to learn and approximate more complex patterns in the data.

**8. What is the main difference between the loss functions used for regression and classification tasks in neural networks?**  
a) Regression uses cross-entropy loss, while classification uses mean squared error (MSE)  
b) Regression uses mean squared error (MSE), while classification typically uses cross-entropy loss  
c) Both tasks use the same loss function, but with different learning rates  
d) Classification uses a linear loss, while regression uses a logistic loss  

**Correct Answer: b) Regression uses mean squared error (MSE), while classification typically uses cross-entropy loss**  
*Explanation:* MSE is used in regression tasks to measure the difference between predicted and actual values, while cross-entropy loss is used in classification to compare predicted probabilities to true labels.

**9. In a classification task with 3 possible output classes, why is the softmax activation function used in the output layer?**  
a) It introduces non-linearity between neurons  
b) It converts logits into probabilities that sum to 1, which is necessary for classification  
c) It initializes the weights in the final layer  
d) It prevents overfitting by applying regularization to the output layer  

**Correct Answer: b) It converts logits into probabilities that sum to 1, which is necessary for classification**  
*Explanation:* Softmax normalizes the output logits into probabilities, which is crucial for multi-class classification tasks where the predicted class is determined based on the highest probability.

**10. How does increasing the number of layers or neurons in a fully connected neural network impact its performance?**  
a) It always improves accuracy with no drawbacks  
b) It increases the model's ability to learn complex patterns but can lead to overfitting if the network is too large  
c) It reduces the risk of overfitting and makes the network more robust  
d) It decreases training time and improves generalization  

**Correct Answer: b) It increases the model's ability to learn complex patterns but can lead to overfitting if the network is too large**
