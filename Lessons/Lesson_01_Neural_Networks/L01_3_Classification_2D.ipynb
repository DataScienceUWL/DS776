## Prepare the Data - Datasets and DataLoaders

We use `sklearn.datasets.make_circles` from Scikit-Learn to create a dataset similar to the one in Neural Network Playground.  

Notice that the randomly selected points are stored as 64 bit floats.  We'll have to change those to 32 bit floats for PyTorch.

```python
X_train, y_train = make_circles(n_samples=1000, factor=0.3, noise=0.2, random_state=123)
print(X_train.dtype, y_train.dtype)
X_test, y_test = make_circles(n_samples=400, factor=0.3, noise=0.2)

train_dataset = TensorDataset(torch.tensor(X_train, dtype=torch.float32), 
                              torch.tensor(y_train, dtype=torch.long))
test_dataset = TensorDataset(torch.tensor(X_test, dtype=torch.float32), 
                             torch.tensor(y_test, dtype=torch.long))

train_loader = DataLoader(train_dataset, batch_size=50, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=50)
```

## Corrected Model

```python
class CirclesClassifier(nn.Module):
    def __init__(self):
        super(CirclesClassifier, self).__init__()
        self.model = nn.Sequential(
            nn.Linear(2, 6),
            nn.Tanh(),
            nn.Linear(6, 2), # changed the number of outputs to match next layer
            nn.Tanh(),
            nn.Linear(2, 2)
        )

    def forward(self, x):
        return self.model(x)

model = CirclesClassifier()
model
```

## Training setup

There isn't much difference between the structure of the neural network we're using here and the one used previously for regression.  Aside from using different numbers of hidden layers and neurons, the main difference is that the output layer for our classification network has two outputs instead of one.  For classification we want want one output for each class.  

Those outputs are converted to probabilities using a softmax function.  However, in PyTorch, the softmax function is incorporated into the loss function and isn't explicitly included in the neural network.  This approach is common because it lead to more numerically efficient and stable calculations.

The other major difference is that we replace the mean squared error loss function for regression with the cross entropy loss used for classification.  Cross entropy loss is generally used for comparing probability distributions.  The value of the cross entropy loss will be minimized when the output class probabilities from the softmax are as close as possible to the class labels (zeros and ones).

Cross entropy can be difficult to interpret, so we usually track other metrics such as accuracy, sensitivity, etc.  Here we show how track accuracy during training, but you easily add other metrics to the `score_funcs` dictionary.

```python
# Cross Entropy Loss (built in sigmoid)
loss_func = nn.CrossEntropyLoss()

# Accuracy metric for monitoring training
score_funcs = {'ACC':accuracy_score}

# determine device
device = get_device()
print(f"Using device: {device}")

# checkpoint filename
ckpt_file = MODELS_PATH / 'L01_CirclesData_CirclesClassifier.pt'
```

## Making Predictions

The visualization above uses model to predict the probabilities that each $(x,y)$ point in a grid is in class 1.  Note the model outputs two logits for each input.  Logits are what the raw output of the last linear layer are called.  Those logits are converted into a pair of probabilities by the softmax function.  Since there are only two outputs the probabilities must sum to one.  We return only the probability that the point belongs to class 1.

```python
def predict_classes(points, model):
    """
    Predicts the class probabilities and labels for a given set of points using a trained model.

    Args:
        points (numpy.ndarray): An Nx2 array of x,y point coordinates.
        model: The trained model used for prediction.

    Returns:
        tuple: A tuple containing two numpy arrays. The first array contains the probabilities of the points belonging to the second class.
               The second array contains the predicted class labels (0 or 1) for the points.
    """
    # Convert numpy array to torch tensor
    tensor_points = torch.tensor(points, dtype=torch.float32)
    
    # Make predictions using the model
    with torch.no_grad():
        logits = model(tensor_points)
        probabilities = torch.softmax(logits, dim=1)
        _, predicted_labels = torch.max(probabilities, dim=1)
    
    # Convert predicted labels to numpy array
    predicted_labels = predicted_labels.numpy()
    
    return probabilities[:, 1].numpy(), predicted_labels
```

Now let's predict the probability and label for a few points:

```python
x = np.array([[1, 0.5], [0.0, 0.0], [-0.6, -0.7]])
probs, labels = predict_classes(x, model)
print(probs, labels)
```

## Regression vs Classification

When it comes to fully-connected networks, there is little difference between classification and regression tasks. The main distinction lies in the output layer of the network.

For regression tasks, the output layer consists of logits, which are the raw outputs of the network. These logits are directly compared to the target values during training.

On the other hand, for classification tasks, the output layer of the network typically includes a function that converts the raw outputs to probabilities. This function is usually a sigmoid or softmax function. In PyTorch, these functions are often incorporated into the loss function itself, rather than being explicitly included in the network. This approach ensures numerically stable and efficient calculations.

In the previous notebook our nonlinear regression output layer had a single output which was passed to a mean squared error.  In this notebook our classification output layer has two outputs which are passed to a cross entropy loss function.  Otherwise the networks we used were largely the same consisting of fully-connected hidden layers and hyperbolic tangent activation functions.
