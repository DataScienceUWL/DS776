Here’s the revised 10-question multiple-choice quiz for Sections 13.1-13.3:

1. **What is the primary advantage of using transfer learning?**  
   - A. It requires no labeled data  
   - B. It allows reuse of a pre-trained model on a new, related task  
   - C. It always results in higher accuracy than training from scratch  
   - D. It requires no modifications to the model’s architecture  
   - **Answer**: B  

---

2. **In transfer learning, what is the term for keeping some weights fixed while training?**  
   - A. Cold start  
   - B. Warm transfer  
   - C. Frozen weights  
   - D. Weight decay  
   - **Answer**: C  

---

3. **Why is transfer learning particularly useful for CNNs in image tasks?**  
   - A. CNNs learn structural similarities that are often transferable across image datasets  
   - B. CNNs require very few parameters  
   - C. CNNs are only useful for large datasets  
   - D. CNNs do not use weights that need adjusting  
   - **Answer**: A  

---

4. **What is the main purpose of replacing or modifying layers in a pre-trained model for transfer learning?**  
   - A. To increase the model size  
   - B. To adapt the model for the new target classes  
   - C. To reduce training time  
   - D. To freeze more weights  
   - **Answer**: B  

---

5. **Which of the following best describes the "source domain" in transfer learning?**  
   - A. The smaller dataset used for fine-tuning  
   - B. The domain from which the pre-trained model’s knowledge is transferred  
   - C. The data that is specific to the new target task  
   - D. The set of frozen layers in the model  
   - **Answer**: B  

---

6. **In transfer learning, what is a ‘warm start’?**  
   - A. Using a small dataset before the main training  
   - B. Training with partial data  
   - C. Fine-tuning weights that are not frozen  
   - D. Starting training without any pre-trained weights  
   - **Answer**: C  

---

7. **What does the term ‘target domain’ refer to in transfer learning?**  
   - A. The pre-trained dataset  
   - B. The original task the model was trained on  
   - C. The new task or dataset for which the model is being adapted  
   - D. The set of frozen layers in the model  
   - **Answer**: C  

---

8. **Why might fewer training epochs be needed in transfer learning compared to training from scratch?**  
   - A. The model has already learned generalized features from a larger dataset  
   - B. Fewer training samples are needed  
   - C. Pre-trained models do not require backpropagation  
   - D. The model has no need for further optimization  
   - **Answer**: A  

---

9. **Which type of layers are commonly adjusted or replaced when applying transfer learning to a CNN?**  
   - A. Initial convolutional layers  
   - B. Fully connected (dense) layers at the end of the network  
   - C. Pooling layers  
   - D. Activation layers  
   - **Answer**: B  

---

10. **What benefit does transfer learning provide when there is limited labeled data for the target task?**  
    - A. It reduces overfitting by freezing most layers  
    - B. It enables training without labels  
    - C. It allows for generalization using pre-learned features  
    - D. It skips data preprocessing steps  
    - **Answer**: C  

---
